```python
import pandas as pd
# reading
url = "https://drive.google.com/file/d/1xN6hNR23Yi5AQ-XAlFiPCSzIibMtnNOV/view?usp=sharing"#Earlier removed 22 values from SalePrice, 5 red columns
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
traindata =  pd.read_csv(path)
traindata.shape
traindata.drop(['Id','Condition2','SaleType','Heating','RoofMatl','Street'],inplace=True,axis=1)#Totally 11 columns, 22 rows are removed
traindata.info()
type(traindata)
traindata.shape

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split#Do not forget to split later
X = traindata.drop(columns=['SalePrice'])
y = traindata['SalePrice']
X.info()


```

```python
# selecting non-numerical columns
X_cat = X.select_dtypes(exclude="number")
from sklearn.impute import SimpleImputer

# defining the imputer to use "N_A" as replacement value
data_imputer = SimpleImputer(strategy="constant", 
                            fill_value="N_A")

# fitting the imputer
data_imputer.fit(X_cat)

# transforming the data & keeping it as a DataFrame
X_cat_imputed = pd.DataFrame(data_imputer.transform(X_cat), 
                             columns=X_cat.columns)
X_cat_imputed.head()
X_cat_imputed.shape
```

```python
# Selecting numerical columns
X_num = X.select_dtypes(include="number")

# Imputing the mean
num_imputer = SimpleImputer(strategy="mean")

# Fitting
num_imputer.fit(X_num)

# Transforming, keeping a DataFrame
X_num_imputed = pd.DataFrame(num_imputer.transform(X_num), 
                             columns=X_num.columns)

X_num_imputed.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

my_scaler = MinMaxScaler()

X_num_scaledarray = my_scaler.fit_transform(X_num_imputed)
X_num_scaled =pd.DataFrame(X_num_scaledarray,columns=X_num_imputed.columns)

X_num_scaled
 
```

```python
# import
from sklearn.preprocessing import OneHotEncoder

# initialize
my_onehot = OneHotEncoder(drop="first")

# fit
my_onehot.fit(X_cat_imputed)

# transform
X_cat_imputed_onehot = my_onehot.transform(X_cat_imputed)

df = pd.DataFrame.sparse.from_spmatrix(X_cat_imputed_onehot)
df.head()



```

```python
X_cat_imputed
colnames = my_onehot.get_feature_names_out(X_cat_imputed.columns)
df.columns = colnames
df.head()


```

```python
data_imputed1 = pd.concat([df, X_num_scaled], axis=1)
data_imputed1

data_imputed1.head(3)
data_imputed1.info()

from sklearn.model_selection import train_test_split
X = data_imputed1
y = y

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8, random_state=23)
X_train.head()
y_test#Finished all corrections. Then deletion and regression

```

```python
# Decision tree
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor() # initialize
tree.fit(X_train, y_train)

# K Nearest neighbors
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=1) # initialize with parameters
neigh.fit(X_train, y_train) # fit to train set

```

```python
from sklearn.metrics import r2_score

tree_pred = tree.predict(X = X_test) # Decision tree
knn_pred = neigh.predict(X = X_test) # K Nearest neighbors


performances = {}

performances["baseline_tree"]= r2_score(y_test, tree_pred)
performances["baseline_knn"] = r2_score(y_test, knn_pred)#Baseline_knn: bad

performances


```

```python
import seaborn as sn
import matplotlib.pyplot as plt

# X_train.drop(columns=["feature_27", "feature_28"], inplace=True)
#fig, ax = plt.subplots(figsize=(18,18))
corrMatrix = X_train.corr().abs()
#sn.heatmap(corrMatrix, annot=True);

# Select upper triangle of correlation matrix# Droping fea
upper = corrMatrix.where(np.triu(np.ones(corrMatrix.shape), k=1).astype(bool))

# Find index of feature columns with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
to_drop

```

```python
# drop cols train set
X_train_selected = X_train.drop(columns=to_drop, )

# drop cols test set
X_test_selected = X_test.drop(columns=to_drop)
```

```python
# Decision tree
tree = DecisionTreeRegressor() # initialize
tree.fit(X_train_selected, y_train)
tree_pred = tree.predict(X = X_test_selected)

# K Nearest neighbors
neigh = KNeighborsRegressor(n_neighbors=1) # initialize with parameters
neigh.fit(X_train_selected, y_train) # fit to train set
knn_pred = neigh.predict(X = X_test_selected)

performances["collinearity_tree"]= r2_score(y_test, tree_pred)
performances["collinearity_knn"] = r2_score(y_test, knn_pred)

performances

```

```python
#ANOVA, CHI2 !!!
from sklearn.feature_selection import SelectKBest, f_regression#F_regression for regression, should be imported

KBest = SelectKBest(score_func=f_regression, k=19)#K - how many features to select. My choice. 19 is the best, but not better than the baseline

# transform train set
train_X_KBest = KBest.fit_transform(X_train, y_train)

# transform test set
test_X_KBest = KBest.transform(X_test)

```

```python
train_X_KBest.shape
```

```python
# Decision tree
tree = DecisionTreeRegressor() # initialize
tree.fit(train_X_KBest, y_train)
tree_pred = tree.predict(X = test_X_KBest)

# K Nearest neighbors
neigh = KNeighborsRegressor(n_neighbors=1) # initialize with parameters
neigh.fit(train_X_KBest, y_train) # fit to train set
knn_pred = neigh.predict(X = test_X_KBest)

performances["KBest_tree"]= r2_score(y_test, tree_pred)
performances["KBest_knn"] = r2_score(y_test, knn_pred)

performances
```
